# Causal Estimation
A crash-course in causal modeling: https://matheusfacure.github.io/python-causality-handbook/04-Graphical-Causal-Models.html

We want to identify what proportion of babies who have jaundice will have G6PD-deficiency.
To do so, we will use Bayes' Theorem:

P(G6PD-Deficiency | jaundice) = P(jaundice | G6PD-Deficiency)*P(G6PD-Deficiency) / P(jaundice)

The prior probability of G6PD-deficiency will be estimated from the current known prevalence of G6PD-deficiency

However, according to the proposed causal diagram, race and sex are confounders of G6PD-deficiency.
Thus, we must control for race and sex (aka, "condition on" in causal modeling terminology) to eliminate
the effect.

```python
# Load necessary 3rd Party Python Packages
import graphviz as gr
import pandas as pd
import numpy as np
```

```python
# Build causal diagram
g = gr.Digraph()

g.edge("race", "G6PD-deficiency")
g.edge("race", "jaundice")

g.edge("sex", "G6PD-deficiency")
g.edge("sex", "jaundice")

g.edge("G6PD-deficiency", "jaundice")

g
```

```python
# Color the nodes we condition on
g.node('race', color='red')
g.node('sex', color='red')

g
```
